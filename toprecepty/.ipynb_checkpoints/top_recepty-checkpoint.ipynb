```python
import pandas as pd
import numpy as np
import plotly.express as px
import pyodbc
import matplotlib as plt
import seaborn as sns
import matplotlib.colors as mcolors
```

```python
query = '''select
                r.nazev as nazev_receptu,
                nh.nazev as nazev_nutricni_hodnota,
                cast(rnh.mnozstvi as float) as mnozstvi,
                rnh.mnozstvi_jednotky
        from
            kucharka.dbo.recept_nutricni_hodnota rnh
        inner join
            kucharka.dbo.recept r
            on (rnh.recept_id = r.id)
        inner join
            kucharka.dbo.nutricni_hodnota nh
            on (rnh.nutricni_hodnota_id = nh.id)
        where
            nh.nazev not like '%–%'
        ;'''

cnxn = pyodbc.connect(driver='{SQL Server}', server='(local)', database='kucharka', trusted_connection='yes')

df = pd.read_sql(query, cnxn)
df.describe()
```

```python
nutrition_list = ['Bílkoviny', 'Cholesterol', 'Energie', 'Sacharidy', 'Sodík', 'Sůl', 'Tuky', 'Vláknina']

nutrition_df = df[df["nazev_nutricni_hodnota"] == 'Energie']
nutrition_df.describe()
```

```python
palette = sns.color_palette()
nutrition_list = ['Bílkoviny', 'Cholesterol', 'Energie', 'Sacharidy', 'Sodík', 'Sůl', 'Tuky', 'Vláknina']

nutrition_df = df[df["nazev_nutricni_hodnota"] == 'Bílkoviny']
  
sns.boxplot(data=nutrition_df, x='mnozstvi', y='nazev_nutricni_hodnota', orient='h', palette=palette)
```

```python
fig, axs = plt.pyplot.subplots(nrows=4, ncols=2)
fig.subplots_adjust(hspace=1, wspace=1)
fig.suptitle('Hrubá data - Statistika nad jednotlivými nutricnimi hodnotami')
nutrition_list = ['Bílkoviny', 'Cholesterol', 'Energie', 'Sacharidy', 'Sodík', 'Sůl', 'Tuky', 'Vláknina']

cnt = 0
for i in range(4):
    for j in range(2):
        nutrition_df = df[df["nazev_nutricni_hodnota"] == nutrition_list[cnt]]
    
        sns.boxplot(data=nutrition_df, x='mnozstvi', y='nazev_nutricni_hodnota', orient='h', palette=palette, ax=axs[i, j]).set(
            xlabel='', 
            ylabel=''
        )
        cnt = cnt + 1
```

```python
fig, axs = plt.pyplot.subplots(nrows=4, ncols=2)
fig.subplots_adjust(hspace=1, wspace=1)
fig.suptitle('Čistá data - Statistika nad jednotlivými nutricnimi hodnotami')
nutrition_list = ['Bílkoviny', 'Cholesterol', 'Energie', 'Sacharidy', 'Sodík', 'Sůl', 'Tuky', 'Vláknina']
quantile_list = [200, 500, 20000, 500, 500, 10, 300, 40]

cnt = 0
nutrition_df_list = []
quantile_list_out = []
for i in range(4):
    for j in range(2):
        
        nutrition_df = df[df["nazev_nutricni_hodnota"] == nutrition_list[cnt]]
        q_low = nutrition_df["mnozstvi"].quantile(0.25)
        #q_hi  = quantile_list[cnt]
        q_hi = nutrition_df["mnozstvi"].quantile(0.75)
        
        clean_nutrition_df = nutrition_df[(nutrition_df["mnozstvi"] < q_hi) & (nutrition_df["mnozstvi"] > q_low)]
    
        sns.boxplot(data=clean_nutrition_df, x='mnozstvi', y='nazev_nutricni_hodnota', orient='h', palette=palette, ax=axs[i, j]).set(
            xlabel='', 
            ylabel=''
        )
        cnt = cnt + 1
        
        quantile_list_out.append((q_low, q_hi))
        nutrition_df_list.append(clean_nutrition_df)
```
